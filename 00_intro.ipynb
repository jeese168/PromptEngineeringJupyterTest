# 提示词工程（Prompt Engineering）入门

基于吴恩达 x OpenAI 提示词工程课程的实践 notebook

## 📚 什么是提示词工程？

在早期使用 LLMs 进行工程开发时，大多数用例都需要针对一次性分类或文本生成任务，**提示工程关注如何编写有效的提示**，在一定程度上为 AI 工程提升了不错的效果。

### 简洁定义

**提示词工程**：通过精心设计的提示词，尝试让大模型输出更好的回答，来提高 AI Agent 的功能。

### 🎯 核心目标

```
模糊的指令 → 不可预测的输出
清晰的提示 → 可靠的结果
```

## 💡 为什么需要提示词工程？

大语言模型就像一个非常聪明但需要明确指示的助手：

- ❌ **糟糕的提示**："帮我写个东西" → 模型不知道写什么
- ✅ **好的提示**："请将以下产品描述改写为吸引年轻人的营销文案，字数100字以内，风格活泼" → 明确的任务和要求

## 🚀 提示词工程的应用场景

1. **内容生成**：文章写作、代码生成、营销文案
2. **信息提取**：从文本中提取结构化数据
3. **任务自动化**：RAG 系统、AI Agent
4. **创意工作**：头脑风暴、设计建议
5. **数据分析**：情感分析、文本分类

## 📖 本课程学习路径

我们将通过一系列实践 notebook 逐步掌握提示词工程：

### Notebook 1: 核心原则
**文件**: `01_principles.ipynb`

学习两大核心原则：
1. **写清楚、具体的 Prompt**
   - 使用分隔符
   - 请求结构化输出
   - 检查条件
   - Few-shot 提示

2. **给模型时间思考**
   - 指定步骤
   - 先独立解决再评估

### Notebook 2: Shot Prompting 策略
**文件**: `02_shot_prompting.ipynb`

对比三种提示策略的效果：
- **Zero-shot**: 不提供示例（适合简单任务）
- **One-shot**: 提供1个示例（性价比最高）
- **Few-shot**: 提供多个示例（复杂任务必须）

### Notebook 3: 输出格式控制
**文件**: `03_output_format.ipynb`

学习如何精确控制模型输出格式：
- XML 格式（推荐用于复杂结构）
- JSON 格式（推荐用于数据交换）
- HTML 格式（推荐用于内容展示）

## 🔍 快速示例：好提示 vs 坏提示

让我们通过一个简单的例子理解提示词工程的重要性。

### ❌ 糟糕的提示（模糊、不具体）

```
写一个产品介绍
```

**问题**：
- 什么产品？
- 写给谁看？
- 多长？
- 什么风格？

**模型可能的输出**：
> "这是一款很棒的产品，功能强大，质量上乘，值得购买。"  
> *(毫无用处的通用描述)*

---

### ✅ 优秀的提示（清晰、具体、有指导）

```
请为以下产品编写营销介绍：

产品：无线蓝牙耳机
特点：
- 主动降噪
- 30小时续航
- 快充功能（充电10分钟使用3小时）

要求：
1. 目标受众：25-35岁职场人士
2. 字数：100-150字
3. 风格：专业但不失温度
4. 突出核心卖点：降噪和续航
5. 使用 XML 格式输出，包含 <title> 和 <description> 标签
```

**模型可能的输出**：
```xml
<product>
  <title>告别嘈杂，专注每一刻</title>
  <description>
    在拥挤的地铁、喧闹的咖啡厅，这款主动降噪耳机为你开启专属静谧空间。
    30小时超长续航陪伴你一周通勤，即使忘记充电也不必担心——
    快充10分钟即可畅听3小时。无论工作汇报还是深夜追剧，
    让声音的每个细节都清晰可闻。这不只是一副耳机，
    更是你掌控环境、专注自我的利器。
  </description>
</product>
```

---

## 🎯 关键差异

| 维度 | 糟糕提示 | 优秀提示 |
|-----|---------|---------|
| **任务描述** | 模糊 | 明确具体 |
| **上下文信息** | 缺失 | 完整提供 |
| **输出要求** | 无 | 详细规定（长度、风格、格式） |
| **可复现性** | 低（每次输出差异大） | 高（输出稳定可控） |
